### Loading the datasets  
This project uses the **Ames Housing Dataset** to create and evaluate a regression model to predict the saleprice of a house.   

**About the dataset:** this dataset is adopted from the Ames, Iowa Assessor’s Office and it contains information from the Ames Assessor’s Office used in computing assessed values for individual residential properties sold in Ames, IA from 2006 to 2010 ([click here for the data documentation](http://jse.amstat.org/v19n3/decock/DataDocumentation.txt)).

### Exploring the Categorical Variables

```python
# review the barcharts and dummify the categorical variables.
# I also combine some the categories, for example
# lot_shape currently have regular and three irregular categories 
# I will only categorized them as regular = 1 and irregular = 0

#for each in categorial_features:
#    train[each].value_counts().plot(kind='bar')
#    plt.xlabel(f'{each}')
#    plt.show();
```

```python
# create 'lot_regular' based on 'lot_shape'

print(train['lot_shape'].value_counts())
# lot_shape currently have regular and three irregular categories 
# I will only categorized them as regular = 1 and irregular = 0

train['lot_regular'] = train['lot_shape'].map(lambda row: 1 if row=='Reg' else 0)
train['lot_regular'].value_counts()
```

```python
# create 'more_than_1story' based on 'house_style'

print(train['house_style'].value_counts())

train['more_than_1story'] = train['house_style'].map(lambda row: 0 if row=='1Story' else 1)
train['more_than_1story'].value_counts()
```

```python
# create 'foundation_PConc' based on 'foundation' 

print(train['foundation'].value_counts())

train['foundation_PConc'] = train['foundation'].map(lambda row: 1 if row=='PConc' else 0)
train['foundation_PConc'].value_counts()
```

```python
# create 'has_bsmt_exposure' based on 'bsmt_exposure'

print(train['bsmt_exposure'].value_counts())

train['has_bsmt_exposure'] = train['bsmt_exposure'].map(lambda row: 0 if row=='No' else 1)
train['has_bsmt_exposure'].value_counts()
```

```python
# create 'bsmtfin_type_GLQ' based on 'bsmtfin_type_1'

print(train['bsmtfin_type_1'].value_counts())

train['bsmtfin_type_GLQ'] = train['bsmtfin_type_1'].map(lambda row: 1 if row=='GLQ' else 0)
train['bsmtfin_type_GLQ'].value_counts()
```

```python
# create 'has_central_air' based on 'central_air'
print(train['central_air'].value_counts())
train['has_central_air'] = train['central_air'].map(lambda row: 1 if row=='Y' else 0)
train['has_central_air'].value_counts()
```

```python
# create 'garage_finish_dum' based on 'garagre_finish'
train['garage_finish'].value_counts()

train['garage_finish_dum'] = train['garage_finish'].map(lambda row: 0 if row=='Unf' else 1)
train['garage_finish_dum'].value_counts()
```

```python
def ex_qual(row):
    row = str(row)
    if row=='Ex':
        return 1
    else: 
        return 0
```

```python
# create 'exter_ex' based on 'exter_qual'

print(train['exter_qual'].value_counts())

train['exter_ex'] = train['exter_qual'].apply(ex_qual)
train['exter_ex'].value_counts()
```

```python
# create 'bsmt_ex' based on 'bsmt_qual'

print(train['bsmt_qual'].value_counts())

train['bsmt_ex'] = train['bsmt_qual'].apply(ex_qual)
train['bsmt_ex'].value_counts()
```

```python
# create 'heating_ex' based on 'heating_qc'

print(train['heating_qc'].value_counts())

train['heating_ex'] = train['heating_qc'].apply(ex_qual)
train['heating_ex'].value_counts()
```

```python
# 'kitchen_qual'

print(train['kitchen_qual'].value_counts())

train['kitchen_ex'] = train['kitchen_qual'].apply(ex_qual)
train['kitchen_ex'].value_counts()
```

```python
# After the dummification, the new categorical features are: 
categorical_features = ['lot_regular', 'more_than_1story', 'exter_ex', 'foundation_PConc', 'bsmt_ex', 
                       'has_bsmt_exposure', 'heating_ex', 'kitchen_ex', 'garage_finish_dum', 
                       'has_central_air', 'bsmtfin_type_GLQ']
```

```python
# define a categorical_pre function to preprocess the selected categorical feature all at once
# this would be helpful when we process the testing data

def ex_qual(row):
    row = str(row)
    if row=='Ex':
        return 1
    else: 
        return 0

def categorical_pre(df):
    # create 'lot_regular' based on 'lot_shape'j
    df['lot_regular'] = df['lot_shape'].map(lambda row: 1 if row=='Reg' else 0)
    # create 'more_than_1story' based on 'house_style'
    df['more_than_1story'] = df['house_style'].map(lambda row: 0 if row=='1Story' else 1)
    # create 'foundation_PConc' based on 'foundation' 
    df['foundation_PConc'] = df['foundation'].map(lambda row: 1 if row=='PConc' else 0)
    # create 'has_bsmt_exposure' based on 'bsmt_exposure'
    df['has_bsmt_exposure'] = df['bsmt_exposure'].map(lambda row: 0 if row=='No' else 1)
    # create 'bsmtfin_type_GLQ' based on 'bsmtfin_type_1'
    df['bsmtfin_type_GLQ'] = df['bsmtfin_type_1'].map(lambda row: 1 if row=='GLQ' else 0)
    # create 'has_central_air' based on 'central_air'
    df['has_central_air'] = df['central_air'].map(lambda row: 1 if row=='Y' else 0)
    # create 'garage_finish_dum' based on 'garagre_finish'
    df['garage_finish_dum'] = df['garage_finish'].map(lambda row: 0 if row=='Unf' else 1)
    # create 'exter_ex' based on 'exter_qual'
    df['exter_ex'] = df['exter_qual'].apply(ex_qual)
    # create 'bsmt_ex' based on 'bsmt_qual'
    df['bsmt_ex'] = df['bsmt_qual'].apply(ex_qual)
    # create 'heating_ex' based on 'heating_qc'
    df['heating_ex'] = df['heating_qc'].apply(ex_qual)
    # create 'kitchen_ex' based on 'kitchen_qual'
    df['kitchen_ex'] = df['kitchen_qual'].apply(ex_qual)
```
